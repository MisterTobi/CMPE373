```python
import numpy as np # for random 
import matplotlib.pyplot as plt # for visual effects
```

```python
class Ecoman():
    def __init__(self, ID, money = 10):
        self.ID = ID
        self.budget = money
        
    def take(self, amount):
        self.budget = self.budget + amount
        
    def give_original(self, amount = 1):
        if self.budget >= amount:
            self.budget -= amount
            return amount
        return 0
    
    def give(self, amount = 1):
        self.budget -= amount
        return amount
```

```python
class Ecomania():
    def __init__(self, N = 10):
        self.N = N
        self.population = [Ecoman(i) for i in range(self.N)]
        
    def exchange(self):
        taker_index, giver_index = np.random.choice(self.N, size = 2, replace = False)
        
        amount = self.population[giver_index].give()
        self.population[taker_index].take(amount)
        
    def run(self, G = 100):
        # Do 100 transactions/exchange of money
        for i in range(G):
            self.exchange()
```

# Initial Virtual World

```python
world = Ecomania(N = 100)
```

```python
budgets = [e.budget for e in world.population]
```

```python
plt.hist(budgets)
plt.xlabel('Budget of the Agent')
plt.ylabel('Number of agents')
```

## After 1 exchange

```python
world.exchange()
```

```python
budgets = [e.budget for e in world.population]
```

```python
plt.hist(budgets)
plt.xlabel('Budget of the Agent')
plt.ylabel('Number of agents')
plt.grid()
```
